```python
import pandas as pd
import requests
```

```python
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
```

```python
data=requests.get(url)
data
```

```python
#create json data variable
jdata=data.json()
jdata[0]
```

```python
len(jdata)
```

```python
#create dataframe
df = pd.DataFrame(jdata, columns=['id', 'url', 'repository_url', 'labels_url', 'user'])
df.head()
```

```python
user=df['user']
user
```

```python
user[0].keys()
```

```python
#create columns in dataframe from user dictionary values

for i in user[0].keys():
    lst=[]
    for j in range(len(user)):
        lst.append(user[j][i])
    
    if i=='id':
        df['user_id']=lst
    elif i=='url':
        df['user_url']=lst
    else:
        df[i]=lst
        
```
