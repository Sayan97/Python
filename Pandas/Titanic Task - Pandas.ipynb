## 10. Seperate Cabin column into num and char 

```python
df.head()
```

```python
num=[]
alpha=[]
for i in df['Cabin']:
    s1=''
    s2=''
    if type(i)==str:
        for char in i:
            if char.isalpha()==True:
                s1+=char
            else:
                s2+=char
                
    num.append(s2)
    alpha.append(s1)   
```

```python
num[:5]
```

```python
alpha[:5]
```

```python
df['Cabin_Char']=alpha
df['Cabin_Num']=num
```
